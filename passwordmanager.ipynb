```python
from tkinter import *
import tkinter as tk
import cx_Oracle
from tkinter import messagebox
from tkinter.ttk import *
from PIL import ImageTk, Image
from tkinter import PhotoImage
from random import randint
import hashlib
window=tk.Tk()
window.title("Signup/login page")
window.configure(bg="white")
window.geometry("800x600")
```

```python
def clear():
    list = window.grid_slaves()
    for l in list:
        l.destroy()
```

```python
p1=0
c=0
d=0
from random import randint
def resultcomp(i1,username):
    global p1,c,d
    do=0
    po=0
    lo=0
    i2=randint(1,3)
    if i2==1:
        i2='rock'
    if i2==2:
        i2='paper'
    if i2==3:
        i2='scissors'
        
    if (i1=='rock' and i2=='scissors')or(i1=='paper' and i2=='rock') or (i1=='scissors' and i2=='paper'):
        p1+=1
        po+=1
        lb=Label(window,text="player 1 wins!",font=("Arial Bold",14)).grid(column=0,row=4)
        
    elif (i2=='paper' and i1=='rock') or(i2=='scissors' and i1=='paper')or(i2=='rock' and i1=='scissors'):
        c+=1
        lo+=1
        lb=Label(window,text="Computer wins!",font=("Arial Bold",14)).grid(column=0,row=4)
        
    else:
        lb=Label(window,text="      draw match!   ",font=("Arial Bold",14)).grid(column=0,row=4)
        do+=1
        d=d+1 
    
    comp=Label(window,text="  The computer chose: {}   ".format(i2),font=("Arial Bold",14)).grid(column=0,row=5)
    w=Label(window,text="Wins for computer: {}".format(c),font=("Arial Bold",14)).grid(column=0,row=6)
    p=Label(window,text="Wins for player 1: {}".format(p1),font=("Arial Bold",14)).grid(column=0,row=7)
    draw=Label(window,text="Draws: {}".format(d),font=("Arial Bold",14)).grid(column=0,row=8)
    
    db=cx_Oracle.connect('system/talha123@localhost')
    cursor=db.cursor()
    cursor.prepare('update pwordmanager set wins=wins+:w where username=:userr')
    cursor.execute(None,{'w':po,'userr':username})
    cursor.prepare('update pwordmanager set draws=draws+:dr where username=:userr')
    cursor.execute(None,{'dr':do,'userr':username})
    cursor.prepare('update pwordmanager set losses=losses+:l where username=:userr')
    cursor.execute(None,{'l':lo,'userr':username})
    db.commit()
```

```python
def middle(i1,username):
    clear()
    lbl2=tk.Label(window,text="Enter choice for player 2",font=("Arial Bold",20),fg="green",bg="white")
    lbl2.grid(column=0,row=0)
    l=tk.Button(window,text="rock",command=lambda:resultplayer2(i1,'rock',username),font=("Arial Bold",14),fg="white",bg="light blue")
    l.grid(column=0,row=1)
    l2=tk.Button(window,text="paper",command=lambda:resultplayer2(i1,'paper',username),font=("Arial Bold",14),fg="white",bg="blue")
    l2.grid(column=0,row=2)
    l3=tk.Button(window,text="scissors",command=lambda:resultplayer2(i1,'scissors',username),font=("Arial Bold",14),fg="white",bg="dark blue")
    l3.grid(column=0,row=3)
```

```python
def player2start(username):
    clear()
    lbl=tk.Label(window,text="Enter choice for player 1",font=("Arial Bold",20),bg="white",fg="green")
    lbl.grid(column=0,row=0)
        
    l=tk.Button(window,text="rock",command=lambda:middle('rock',username),font=("Arial Bold",14),fg="white",bg="light blue")
    l.grid(column=0,row=1)
    l2=tk.Button(window,text="paper",command=lambda:middle('paper',username),font=("Arial Bold",14),fg="white",bg="blue")
    l2.grid(column=0,row=2)
    l3=tk.Button(window,text="scissors",command=lambda:middle('scissors',username),font=("Arial Bold",14),fg="white",bg="dark blue")
    l3.grid(column=0,row=3)
```

```python
def computer(username):
    clear()
    rad1 = tk.Button(window,text='rock',command=lambda:resultcomp('rock',username),font=30)
    rad2 =tk.Button(window,text='paper',command=lambda:resultcomp('paper',username),bg="orange",fg="white",font=30)
    rad3 = tk.Button(window,text='scissors',command=lambda:resultcomp('scissors',username),font=30,bg="blue",fg="white" )
    lbl=Label(window,text="Please make your selection",font=("Arial Bold",30))
    lbl.grid(column=0,row=0)
    
    rad1.grid(column=0,row=1)
 
    rad2.grid(column=0,row=2)
 
    rad3.grid(column=0,row=3)
    
```

```python
def clicked():
    messagebox.showinfo('Instructions','You are player 1\nChoose whether you want to play against player 2 or computer\nrock beats scissors\npaper beats rock\nscissors beats paper')
```

```python
def rockpaperscissors(username):
    clear()
    menu = Menu(window) 
    window.config(menu=menu) 
    filemenu = Menu(menu) 
    menu.add_cascade(label='Start', menu=filemenu)
    filemenu.add_command(label='New game against player 2',command=lambda:player2start(username)) 
    filemenu.add_command(label='New game against computer',command=lambda:computer(username))
    filemenu.add_command(label='Home Page',command=lambda:rockpaperscissors(username))
    filemenu.add_separator() 
    filemenu.add_command(label='Logout', command=start)
    helpmenu = Menu(menu) 
    menu.add_cascade(label='Help', menu=helpmenu) 
    helpmenu.add_command(label='Instructions',command=clicked) 
    db=cx_Oracle.connect('system/talha123@localhost')
    cursor=db.cursor()
    s="select * from pwordmanager"
    cursor.execute(s)
    results=cursor.fetchall()
    db.commit()
    b=False
    for row in results:
        if row[0]==username:
            tk.Label(window,text="Welcome to the rock-paper-scissors gameshow!",font=("Arial Bold",18),fg="red",bg="white").grid(column=0,row=0)
            tk.Label(window,text="Your current record is as follows:",font=("Arial Bold",14),fg="black",bg="white").grid(column=0,row=1)
            tk.Label(window,text="Wins:",font=("Arial Bold",12),fg="black",bg="white").grid(column=0,row=2)
            tk.Label(window,text=row[4],font=("Arial Bold",12),fg="black",bg="white").grid(column=1,row=2)
            tk.Label(window,text="Losses:",font=("Arial Bold",12),fg="black",bg="white").grid(column=0,row=3)
            tk.Label(window,text=row[5],font=("Arial Bold",12),fg="black",bg="white").grid(column=1,row=3)
            tk.Label(window,text="Draws:",font=("Arial Bold",12),fg="black",bg="white").grid(column=0,row=4)
            tk.Label(window,text=row[6],font=("Arial Bold",12),fg="black",bg="white").grid(column=1,row=4)
        
            load = Image.open("gameshowhost.png")
            render = ImageTk.PhotoImage(load)
            img = Label(image=render)
            img.image = render
            img.grid(column=0, row=5)
            load = Image.open("rockpaper.jpg")
            render = ImageTk.PhotoImage(load)
            img = Label(image=render)
            img.image = render
            img.grid(column=1, row=5)

```

```python
def insert(user1,passwd,secure1,question):
    db=cx_Oracle.connect('system/talha123@localhost')
    cursor=db.cursor()
    s="select * from pwordmanager"
    cursor.execute(s)
    results=cursor.fetchall()
    db.commit()
    b=False
    for row in results:
        if row[0]==user1:
            b=True
            tk.Label(window,text="Username is taken").grid(column=1,row=3)
    if b==False:
        cursor.prepare('insert into pwordmanager values(:user12,:pwd,:secu,:ques,0,0,0)')
        cursor.execute(None,{'user12':user1,'pwd':passwd,'secu':secure1,'ques':question})
        db.commit()
        clear()
        tk.Button(window,text="Successful signup-Go back to start page",command=start,font=("Arial Bold",14),fg="white",bg="blue").grid(column=1,row=0)
```

```python
def signup():
    global question
    clear()
    db=cx_Oracle.connect('system/talha123@localhost')
    cursor=db.cursor()
    tk.Label(window,text="Enter username ",font=("Arial Bold",14),fg="black",bg="white").grid(column=0,row=0)
    username=tk.Entry(window,width=20)
    username.grid(column=1,row=0)
    tk.Label(window,text="Enter password",font=("Arial Bold",14),fg="black",bg="white").grid(column=0,row=1)
    password=tk.Entry(window,width=20)
    password.grid(column=1,row=1)
    
    l=["What was your favourite tv show in your childhood?","What street did you grow up on?","Who is your favourite celebrity?"]
    question=l[randint(0,2)]
    tk.Label(window,text="Security question:",font=("Arial Bold",14),fg="black",bg="white").grid(column=0,row=2)
    tk.Label(window,text=question,font=("Arial Bold",12),fg="black",bg="white").grid(column=1,row=2)
    security=tk.Entry(window,width=20)
    security.grid(column=2,row=2)
    submit=tk.Button(window,text="Submit",command=lambda:insert(username.get(),hashlib.sha256((password.get()).encode()).hexdigest(),security.get(),question),font=("Arial Bold",14),fg="white",bg="blue").grid(column=0,row=3)  
    
    
```

```python
def newpwd(user1,pword):
    #print(user1,pword)
    db=cx_Oracle.connect('system/talha123@localhost')
    cursor=db.cursor()
    encrypt=hashlib.sha256(pword.encode()).hexdigest()
    cursor.prepare('update pwordmanager set passwordd=:pwd where username=:userr')
    cursor.execute(None,{'pwd':encrypt,'userr':user1})
    db.commit()
    start()
    
```

```python
def mid(securityanswer):
    db=cx_Oracle.connect('system/talha123@localhost')
    cursor=db.cursor()
    s="select * from pwordmanager"
    cursor.execute(s)
    results=cursor.fetchall()
    db.commit()
    b=False
    for row in results:
        if securityanswer==row[2]:              #matching with security answer
                usern=row[0]
                clear()
                b=True
                tk.Label(window,text="Set new password",font=("Arial Bold",14),fg="black",bg="white").grid(column=0,row=0)
                txt=tk.Entry(window,width=20)
                txt.grid(column=1,row=0)
                tk.Button(window,text=" Submit and return to start page ",command=lambda:newpwd(usern,txt.get()),font=("Arial Bold",14),fg="black",bg="blue").grid(column=1,row=1)
    if b==False:
        clear()
        tk.Label(window,text="Wrong answer",font=("Arial Bold",14),fg="black",bg="white").grid(column=0,row=0)
        tk.Button(window,text="Go back to start page",command=start,font=("Arial Bold",14),fg="white",bg="blue").grid(column=1,row=0)
```

```python
def forgot(username):
    clear()
    db=cx_Oracle.connect('system/talha123@localhost')
    cursor=db.cursor()
    s="select * from pwordmanager"
    cursor.execute(s)
    results=cursor.fetchall()
    db.commit()
    b=False
    for row in results:
        if row[0]==username:
            tk.Label(window,text=row[3],font=("Arial Bold",14),fg="black",bg="white").grid(column=0,row=0)
            txt=tk.Entry(window,width=20)
            txt.grid(column=1,row=0)
            tk.Button(window,text="Submit",command=lambda:mid(txt.get()),font=("Arial Bold",14),fg="white",bg="blue").grid(column=1,row=1)
            
        
```

```python
def passwordcheck(password,username):
    db=cx_Oracle.connect('system/talha123@localhost')
    cursor=db.cursor()
    e=hashlib.sha256(password.encode()).hexdigest()
    s="select * from pwordmanager"
    cursor.execute(s)
    results=cursor.fetchall()
    db.commit()
    b=False
    for row in results:
        if row[0]==username and row[1]==e:
            rockpaperscissors(username)
            b=True
    if b==False:
        tk.Label(window,text=" Unsuccessful login!    ",font=("Arial Bold",14),fg="black",bg="white").grid(column=2,row=0)
        tk.Button(window,text="Forgot password?",command=lambda:forgot(username),font=("Arial Bold",14),fg="white",bg="blue").grid(column=1,row=2)
```

```python

def usernamecheck(username):
    clear()
    db=cx_Oracle.connect('system/talha123@localhost')
    cursor=db.cursor()
    s="select * from pwordmanager"
    cursor.execute(s)
    results=cursor.fetchall()
    db.commit()
    b=False
    for row in results:
        if row[0]==username:
            tk.Label(window,text="Enter password: ",font=("Arial Bold",16),fg="black",bg="white").grid(column=0,row=0)
            txt=tk.Entry(window,width=15)
            txt.grid(column=1,row=0)
            for i in range(4):
                 tk.Label(window,text="",bg="white").grid(column=1,row=i+1)
            load = Image.open("loading.jpg")
            render = ImageTk.PhotoImage(load)
            img = Label(image=render)
            img.image = render
            img.grid(column=1, row=11)
            b=True
            login=tk.Button(window,text="Login                  ",command=lambda:passwordcheck(txt.get(),username),font=("Arial Bold",14),fg="white",bg="blue").grid(column=1,row=1)
    if b==False:
        tk.Label(window,text="Wrong username  ",font=("Arial Bold",20),fg="black",bg="white").grid(column=2,row=0)
        tk.Button(window,text="Go back to start page",command=start,font=("Arial Bold",20),fg="white",bg="red").grid(column=2,row=1)
```

```python
def start():
    emptyMenu = Menu(window)
    window.config(menu=emptyMenu)
    clear()
    lbl=tk.Label(window,text="Enter username to login:",font=("Arial Bold",16),fg="black",bg="white").grid(column=0,row=0)
    txt=tk.Entry(window,width=20)
    txt.grid(column=1,row=0)
    tk.Label(window,text="",bg="white").grid(column=1,row=1)
    pwd=tk.Button(window,text="Proceed to enter password",command=lambda:usernamecheck(txt.get()),font=("Arial Bold",14),fg="white",bg="blue").grid(column=1,row=2)
    tk.Label(window,text="",bg="white").grid(column=1,row=3)
    sign=tk.Button(window,text="Signup",command=signup,font=("Arial Bold",14),fg="white",bg="blue").grid(column=1,row=4)
    tk.Label(window,text="",bg="white").grid(column=1,row=5)
    exit=tk.Button(window,text="Exit",command=window.destroy,font=("Arial Bold",14),fg="white",bg="blue").grid(column=1,row=6)
    load = Image.open("curtains.jpg")
    render = ImageTk.PhotoImage(load)
    img = Label(image=render)
    img.image = render
    img.grid(column=1, row=7)
    window.mainloop()
```
